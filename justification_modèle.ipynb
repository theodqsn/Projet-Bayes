Expliquons les choix du modèle (structure, lois a priori, loi composées)

Modèle mixture gaussienne :    

Le fait de choisir une mixture de deux lois avec un mode est très pertinent au vu de la distribution reconstruite avec les noyeaux (cf validation_du_modele_ME.qmd).

Le fait que ces lois unimodales soient des Gaussiennes peut se justifier avec les arguments suivants :    
-le modèle gaussien est simple (peu de paramètres, passe bien à la conjugaison) \
-on a ici affaire à des répartitions de population, et c'est un modèle classique pour ce genre de problèmes

Lambdas :

pour $\lambda_1$ et $\lambda_2$, on choisit une loi gaussienne centrée sur les moyennes "observées" avec EM, et avec un écart type de $10^{-6}$. On peut expliquer ce choix de la manière suivante :    

-la moyenne calculé avec l'algorithme EM est le meilleure choix sans plus d'informations


-Les variables aléatoires donnant, pour chaque singe, la longueur d'onde peut être décomposé en deux variables aléatoires indépendantes (la première donnant la longueur d'onde (de moyenne $\lambda_1$ pour le groupe 1 et $\lambda_2$ pour le groupe 2), et la seconde qui ajoute une incertitude sur la mesure (centrée autour de la valeur réelle). Ainsi, on voit bien que, pour le groupe $i$, $\lambda_i$ sera une moyenne empirique de variables aléatoires iid, donc est bien représenté par une gaussienne

Alpha :

On cherche une distribution pour une loi de proba d'une variable aléatoire qui répartit les données en deux groupes. Une loi de Dirichlet est donc l'outil privilégié

De plus, on a pas d'information a priori sur les proportions de singes dans chaque groupe, donc on choisit une loi de Dirichlet (1,1)

Tau

On suppose a priori que $\tau$ suive une loi $\Gamma(0.001, 0.001)$, ce qui permet :    
-De faire en sorte d'avoir de petites valeurs pour tau, (la denstité décroit très rapidement depuis 0) \
-D'avoir une loi qui se conjugue bien

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

a = 0.001
b = 0.001

val=0.0001
x = np.linspace(val/100, 0.0001, 1000)

y = stats.gamma.pdf(x, a=a, scale=1/b)

plt.plot(x, y)
plt.title('Densité de la loi Gamma(0.001, 0.001)')
plt.xlabel('x')
plt.ylabel('Densité')
plt.show()
```
